```python

def getRelatedLemmas(target_synset):
  noun_relations=["hypernyms","instance_hypernyms",
                "hyponyms","instance_hyponyms",
                "member_holonyms","part_holonyms", "substance_holonyms",
                "member_meronyms","part_meronyms","substance_meronyms",
                "attributes",
                "topic_domains", "in_topic_domains",
                "usage_domains", "in_usage_domains",
                "region_domains", "in_region_domains",
                ]
  syns_list=[]
  for rel in noun_relations:
    syns_list+= getattr(target_synset,rel)()
  print("Synset list:", syns_list)
  lemmas=[]
  for syns in syns_list:
    for lem in syns.lemma_names():
      lemmas+=[lem]
  return lemmas

def relationSimilarities(word, target_synset):
  print("Definitie: ", target_synset.definition())

  lemmaList=getRelatedLemmas(target_synset)
  for lem in lemmaList:
    if lem in model:
      print(f"Similarity: {word}, {lem}, {model.similarity(word,lem)}")


word="forest"
syns_cat=wn.synsets(word)[0]
relationSimilarities(word,syns_cat)

word="forest"
syns_forest=wn.synsets(word)[0]
relationSimilarities(word,syns_forest)
```
